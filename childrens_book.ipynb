### "How do I become the next Dr. Seuss?"

The goal of this project is to read through a repository of Dr. Seuss books and see how we can emulate his writing style. To do this, we will be using Natural Language Processing (NLP) to graph the frequency distribution of the most popular nouns, verbs, and adjectives used in Seuss books.

```python
# Read text file 
def read_file():
    text = ''
    with open('_REPLACE_THIS_', 'rt') as file_in:
        for line in file_in:
            text = text + line
    return text

# In this case, I have a text file of 6 Dr. Seuss books with a bonus Seussical script 
content = read_file()
```

```python
# Let's see the first 104 characters in the text file we just read
print(content[0:104])
```

```python
# Using Python's Natural Language Toolkit
from nltk import word_tokenize
from nltk.corpus import stopwords 

# Input: "A Cat in the Hat"
# Output: ["Cat", "Hat"] 
def tokenize_stop_words(content):
    # Tokenizing = splitting strings into a list of tokens
    word_tokens = word_tokenize(content) 
    
    # Removing stop words -- commonly used words we can ignore since we aren't interested in knowing
    # how many times Dr. Seuss used "a" or "the" for example
    stop_words = set(stopwords.words('english')) 
    
    filtered = [] 

    for w in word_tokens: 
        if w not in stop_words: 
            filtered.append(w)
    return filtered

tokenized = tokenize_stop_words(content)
```

```python
# Let's see the first 104 tokenized characters
# Notice we see "more" since we took out all the stop words!
print(tokenized[1:104])
```

```python
from nltk import pos_tag

# pos_tag classifies words into their parts of speech
# For example, "Cat" will be tagged as a Proper Noun (NNP)
categorized = pos_tag(tokenized)
```

```python
# Again, checking that the tags look right
print(categorized[0:8])
```

```python
verbs = []
nouns = []
adjectives = []

# Save the verbs, nouns and adjectives so we can analyze them separately
for item in categorized:
        if item[1].startswith('VB'):
            verbs.append(item[0])
        elif item[1].startswith('NN'):
            nouns.append(item[0])
        elif item[1].startswith('JJ'):
            adjectives.append(item[0])
```

```python
# Graphing the frequency distributions of every verb
from nltk import FreqDist

wtPairs = verbs
wfd = FreqDist (verbs)
wfd.plot(30)
```

```python
# Graphing the frequency distributions of every noun
wtPairs = nouns
wfd = FreqDist(nouns)
wfd.plot(30)
```

```python
# Graphing the frequency distributions of every adjective
wtPairs = adjectives
wfd = FreqDist(adjectives)
wfd.plot(30)
```
